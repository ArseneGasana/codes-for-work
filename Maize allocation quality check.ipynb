## Check if maize bundle qty is equal to allocated bundles qty

### 1. Get data on maize bundle quantities per site

```python
#Group sites to show their total maize bundle quantites
MaizeBundle = PreAllocDataMaizeW1W2.groupby('SiteUID')['2021A_Maize kg'].sum().reset_index()

#Merge maize bundle qty with other bundles qty
SiteLevelMaizeCheck = pd.merge(MaizeBundle,FinalMaizeAllocation, on='SiteUID', how='outer')

#Check if maize bundle qty is equal to other bundles qty
SiteLevelMaizeCheck['Check']=SiteLevelMaizeCheck['2021A_Maize kg']==SiteLevelMaizeCheck['Quantity']

#Sites without maize bundle, may be they are from other weeks
SiteLevelMaizeCheck[SiteLevelMaizeCheck['2021A_Maize kg'].isnull()]

#Sites with maize bundle
SiteLevelMaizeCheckNotNull = SiteLevelMaizeCheck[SiteLevelMaizeCheck['2021A_Maize kg'].notnull()]

SiteLevelMaizeqtyCheck = SiteLevelMaizeCheckNotNull[SiteLevelMaizeCheckNotNull['Check']==False]

print('Number of sites where maize bundle qty is different from other bundles qty:', len(SiteLevelMaizeqtyCheck['SiteUID']))
```

## Client level check is Maize bundle qty equal to new allocated qty?

### 1. Get client level other maize bundle dataframe

```python
#Get a client ID
MaizeOnlyDf['ClientID'] = MaizeOnlyDf['DistrictName']+MaizeOnlyDf['OAFID'].astype(str)
SumOfMaizeAllocatedPerClient = MaizeOnlyDf.groupby('ClientID')[['FirstName','LastName','2021A_Pan 53 kg', '2021A_Pan 691 kg','2021A_SC 403 kg', '2021A_SC 637 kg','2021A_WH 403 kg', '2021A_WH 505 kg',
       '2021A_WH 605 kg']].sum().reset_index()
MeltingSumOfMaizeClient = SumOfMaizeAllocatedPerClient.melt(['ClientID'],['2021A_Pan 53 kg', '2021A_Pan 691 kg','2021A_SC 403 kg', '2021A_SC 637 kg','2021A_WH 403 kg', '2021A_WH 505 kg',
       '2021A_WH 605 kg'],'Input Name','Quantity')
FinalMaizeAllocationClient = MeltingSumOfMaizeClient[MeltingSumOfMaizeClient['Quantity']>0]

#Final output
FinalMaizeAllocationClient.head()
```

### 2. Get client level  maize bundles quantity

```python
PreAllocDataMaizeW1W2['ClientID'] = PreAllocDataMaizeW1W2['DistrictName']+PreAllocDataMaizeW1W2['OAFID'].astype(str)
MaizeBundleClient = PreAllocDataMaizeW1W2.groupby(['SiteName','ClientID'])['2021A_Maize kg'].sum().reset_index()
WithMaize = MaizeBundleClient[(MaizeBundleClient['2021A_Maize kg']>0)&(MaizeBundleClient['2021A_Maize kg'].notnull())]

#Compare both data frame
ClientLevelMaizeCheck = pd.merge(WithMaize,FinalMaizeAllocationClient, on='ClientID', how='outer')
ClientLevelMaizeCheck['Check']=ClientLevelMaizeCheck['2021A_Maize kg']==ClientLevelMaizeCheck['Quantity']
#ClientLevelMaizeCheck[ClientLevelMaizeCheck['2021A_Maize kg'].isnull()]
x = ClientLevelMaizeCheck[(ClientLevelMaizeCheck['2021A_Maize kg'].notnull())&(ClientLevelMaizeCheck['Check']==False)]
y = x[(x['Input Name'].notnull())&(x['Quantity'].notnull())]
y
```
