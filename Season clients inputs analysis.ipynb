```python
SeasonClientDetailed = pd.read_csv(r'C:\Users\user\Desktop\Analysis\Burera_Musanze Distribution\Season Clients Detailed_20200727-094030.csv')
DataToClean = SeasonClientDetailed[['DistrictName','SiteName','GroupName','LastName','FirstName','OAFID','2021A_Biolite SHS qty','2021A_DAP kg','2021A_GLP SKP 200 qty','2021A_KAR qty','2021A_Maize kg','2021A_NPK 17 kg','2021A_PICS 100KG qty','2021A_PICS 50KG qty','2021A_POV qty','2021A_SHU qty','2021A_TOM qty','2021A_Travertine kg','2021A_TUN qty','2021A_UREA kg']]
```

#### Total orders per district and site

```python
OrdersPerSite = DataToClean.groupby(['DistrictName','SiteName'])[['2021A_Biolite SHS qty','2021A_DAP kg','2021A_GLP SKP 200 qty','2021A_KAR qty','2021A_Maize kg','2021A_NPK 17 kg','2021A_PICS 100KG qty','2021A_PICS 50KG qty','2021A_POV qty','2021A_SHU qty','2021A_TOM qty','2021A_Travertine kg','2021A_TUN qty','2021A_UREA kg']].sum()
OrdersPerSite
```

#### Unique orders

```python
DataToCleanMelt = DataToClean.melt(['DistrictName','SiteName','GroupName', 'LastName','FirstName','OAFID'],['2021A_Biolite SHS qty','2021A_DAP kg','2021A_GLP SKP 200 qty','2021A_KAR qty','2021A_Maize kg','2021A_NPK 17 kg','2021A_PICS 100KG qty','2021A_PICS 50KG qty','2021A_POV qty','2021A_SHU qty','2021A_TOM qty','2021A_Travertine kg','2021A_TUN qty','2021A_UREA kg'], 'Input Name','Quantity')
PositiveOrders = DataToCleanMelt[DataToCleanMelt['Quantity']>0]
PositiveOrders.drop_duplicates(subset='OAFID',keep=False, inplace=True)
UniquesOrdersPerSite = PositiveOrders.groupby(['DistrictName','SiteName','Input Name'])['OAFID'].count().reset_index()
UniqueOrdersPerProd = PositiveOrders.groupby('Input Name')['OAFID']
UniquesOrdersPerSite.rename(columns={'OAFID':'#Clients'}, inplace=True)
UniqueOrdersPerProdDf = UniqueOrdersPerProd.count().to_frame().reset_index().rename(columns={'OAFID':'#Clients'})
```
